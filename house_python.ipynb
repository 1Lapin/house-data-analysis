```python
import pandas as pd
import numpy as np

df = pd.read_csv(r'D:\桌面2.0\数据分析\house_data\house_data_cleaned_for_tableau.csv')


print("数据形状:", df.shape)
print("\n列名:")
print(df.columns.tolist())
print("\n前5行:")
print(df.head())  # 添加print以显示输出
```

```python
# 数据清洗和统计分析
import warnings
warnings.filterwarnings('ignore')

# 基础统计
print("=== 基础统计 ===")
print(f"总房源数: {len(df)}")
print(f"平均总价: {df['total_price'].mean():.2f}万")
print(f"平均单价: {df['unit_price'].mean():.2f}元/平米")
print(f"平均面积: {df['total_area'].mean():.2f}平米")
print(f"平均房龄: {df['house_age'].mean():.1f}年")

# 地区分布
district_stats = df.groupby('district').agg({
    'total_price': ['count', 'mean', 'median'],
    'unit_price': 'mean',
    'total_area': 'mean'
}).round(2)
district_stats.columns = ['房源数', '平均总价', '中位总价', '平均单价', '平均面积']
district_stats = district_stats.sort_values('房源数', ascending=False)
print("\n=== 地区分布 ===")
print(district_stats)

# 房型分布
room_type = df['房型'].value_counts().head(10)
print("\n=== 房型分布 TOP10 ===")
print(room_type)

# 装修情况
print("\n=== 装修情况 ===")
print(df['装修情况'].value_counts())

# 面积段
print("\n=== 面积段 ===")
print(df['area_range'].value_counts())

# 房龄段
print("\n=== 房龄段 ===")
print(df['age_range'].value_counts())

# 楼层
print("\n=== 楼层分布 ===")
print(df['楼层'].value_counts())

# 朝向
print("\n=== 朝向分布 ===")
print(df['朝向'].value_counts().head(10))
```

```python
import matplotlib.pyplot as plt
import matplotlib
import os

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

# 创建输出目录
os.makedirs('output', exist_ok=True)

# 1. 地区均价柱状图
fig, ax = plt.subplots(figsize=(10, 6))
district_avg = df.groupby('district')['unit_price'].mean().sort_values(ascending=True)
colors = plt.cm.viridis(np.linspace(0.2, 0.8, len(district_avg)))
bars = ax.barh(district_avg.index, district_avg.values, color=colors, edgecolor='white', linewidth=1.5)
ax.set_xlabel('平均单价 (元/平米)', fontsize=12)
ax.set_title('各区域平均单价对比', fontsize=16, fontweight='bold', pad=20)
for bar, val in zip(bars, district_avg.values):
    ax.text(val + 200, bar.get_y() + bar.get_height()/2, f'{val:.0f}', va='center', fontsize=10, fontweight='bold')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()
plt.savefig('output/chart1_district_price.png', dpi=150, bbox_inches='tight', facecolor='white')
plt.close()

# 2. 总价分布直方图
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(df['total_price'], bins=50, color='#4ECDC4', edgecolor='white', alpha=0.8)
ax.axvline(df['total_price'].mean(), color='#FF6B6B', linestyle='--', linewidth=2, label=f'均值: {df["total_price"].mean():.1f}万')
ax.axvline(df['total_price'].median(), color='#FFE66D', linestyle='--', linewidth=2, label=f'中位数: {df["total_price"].median():.1f}万')
ax.set_xlabel('总价 (万)', fontsize=12)
ax.set_ylabel('房源数量', fontsize=12)
ax.set_title('房源总价分布', fontsize=16, fontweight='bold', pad=20)
ax.legend(fontsize=11)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()
plt.savefig('output/chart2_price_dist.png', dpi=150, bbox_inches='tight', facecolor='white')
plt.close()

# 3. 面积段饼图
fig, ax = plt.subplots(figsize=(8, 8))
area_counts = df['area_range'].value_counts()
colors_pie = ['#FF6B6B', '#4ECDC4', '#FFE66D']
wedges, texts, autotexts = ax.pie(area_counts.values, labels=area_counts.index, autopct='%1.1f%%', 
                                   colors=colors_pie, explode=(0.02, 0.02, 0.02),
                                   textprops={'fontsize': 12, 'fontweight': 'bold'})
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontsize(13)
ax.set_title('面积段分布', fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.savefig('output/chart3_area_pie.png', dpi=150, bbox_inches='tight', facecolor='white')
plt.close()

# 4. 房龄段柱状图
fig, ax = plt.subplots(figsize=(10, 6))
age_counts = df['age_range'].value_counts()
colors_age = ['#6C5CE7', '#00B894', '#FD79A8', '#E17055']
bars = ax.bar(age_counts.index, age_counts.values, color=colors_age, edgecolor='white', linewidth=2, width=0.6)
ax.set_ylabel('房源数量', fontsize=12)
ax.set_title('房龄段分布', fontsize=16, fontweight='bold', pad=20)
for bar, val in zip(bars, age_counts.values):
    ax.text(bar.get_x() + bar.get_width()/2, val + 10, str(val), ha='center', fontsize=11, fontweight='bold')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()
plt.savefig('output/chart4_age_dist.png', dpi=150, bbox_inches='tight', facecolor='white')
plt.close()

# 5. 装修情况对比
fig, ax = plt.subplots(figsize=(10, 6))
deco_counts = df['装修情况'].value_counts()
colors_deco = ['#A8E6CF', '#DCEDC1', '#FFD3B6', '#FFAAA5']
bars = ax.bar(deco_counts.index, deco_counts.values, color=colors_deco, edgecolor='white', linewidth=2, width=0.6)
ax.set_ylabel('房源数量', fontsize=12)
ax.set_title('装修情况分布', fontsize=16, fontweight='bold', pad=20)
for bar, val in zip(bars, deco_counts.values):
    ax.text(bar.get_x() + bar.get_width()/2, val + 5, str(val), ha='center', fontsize=11, fontweight='bold')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()
plt.savefig('output/chart5_deco_dist.png', dpi=150, bbox_inches='tight', facecolor='white')
plt.close()

# 6. 单价 vs 面积 散点图
fig, ax = plt.subplots(figsize=(10, 7))
scatter = ax.scatter(df['total_area'], df['unit_price'], c=df['total_price'], 
                    cmap='plasma', alpha=0.6, s=30, edgecolors='none')
ax.set_xlabel('总面积 (平米)', fontsize=12)
ax.set_ylabel('单价 (元/平米)', fontsize=12)
ax.set_title('面积-单价散点图 (颜色=总价)', fontsize=16, fontweight='bold', pad=20)
cbar = plt.colorbar(scatter, ax=ax)
cbar.set_label('总价 (万)', fontsize=11)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()
plt.savefig('output/chart6_scatter.png', dpi=150, bbox_inches='tight', facecolor='white')
plt.close()

# 7. 楼层分布
fig, ax = plt.subplots(figsize=(8, 6))
floor_counts = df['楼层'].value_counts()
colors_floor = ['#74B9FF', '#55EFC4', '#FF7675']
bars = ax.bar(floor_counts.index, floor_counts.values, color=colors_floor, edgecolor='white', linewidth=2, width=0.5)
ax.set_ylabel('房源数量', fontsize=12)
ax.set_title('楼层分布', fontsize=16, fontweight='bold', pad=20)
for bar, val in zip(bars, floor_counts.values):
    ax.text(bar.get_x() + bar.get_width()/2, val + 5, str(val), ha='center', fontsize=11, fontweight='bold')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()
plt.savefig('output/chart7_floor.png', dpi=150, bbox_inches='tight', facecolor='white')
plt.close()

# 8. 各区域房源数量
fig, ax = plt.subplots(figsize=(10, 6))
district_counts = df['district'].value_counts()
colors_d = plt.cm.Set3(np.linspace(0, 1, len(district_counts)))
bars = ax.bar(district_counts.index, district_counts.values, color=colors_d, edgecolor='white', linewidth=2)
ax.set_ylabel('房源数量', fontsize=12)
ax.set_title('各区域房源数量', fontsize=16, fontweight='bold', pad=20)
for bar, val in zip(bars, district_counts.values):
    ax.text(bar.get_x() + bar.get_width()/2, val + 3, str(val), ha='center', fontsize=10, fontweight='bold')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('output/chart8_district_count.png', dpi=150, bbox_inches='tight', facecolor='white')
plt.close()

print("所有图表已生成!")
print("当前工作目录:", os.getcwd())
print("图表保存路径:", os.path.abspath('/output'))
```

```python
import base64
import os

# 读取所有生成的图表并转换为base64
chart_files = [
    'chart1_district_price.png',
    'chart2_price_dist.png', 
    'chart3_area_pie.png',
    'chart4_age_dist.png',
    'chart5_deco_dist.png',
    'chart6_scatter.png',
    'chart7_floor.png',
    'chart8_district_count.png'
]

chart_base64 = {}
for f in chart_files:
    path = f'/mnt/agents/output/{f}'
    with open(path, 'rb') as img:
        chart_base64[f] = base64.b64encode(img.read()).decode('utf-8')

print("图表已编码为base64")
print(f"共 {len(chart_base64)} 个图表")
```
